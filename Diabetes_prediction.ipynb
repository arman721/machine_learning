```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

```python
diabetes_data=pd.read_csv("/content/diabetes.csv")
```

```python
X=diabetes_data.drop(columns="Outcome",axis=1)
Y=diabetes_data["Outcome"]
```

```python
scaler=StandardScaler()
scaler.fit(X)
Standarized_data=scaler.transform(X)
X=Standarized_data
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
modal=svm.SVC(kernel="linear")
modal.fit(X_train,Y_train)
train_prediction=modal.predict(X_train)
train_accuracy=accuracy_score(train_prediction,Y_train)
print("The accuracy score for training modal is",train_accuracy)

```

```python
test_prediction=modal.predict(X_test)
test_accuracy=accuracy_score(test_prediction,Y_test)
print("The accuracy score for testing modal is",test_accuracy)
```

```python
input_data=(5,166,72,19,175,25.8,0.587,51)
input_data_asarray=np.asarray(input_data)
input_data_asarray_reshaped=input_data_asarray.reshape(1,-1)
standard_input=scaler.transform(input_data_asarray_reshaped)
prediction=modal.predict(standard_input)
print(prediction)

if(prediction[0]==0):
  print("The person is not diabetes")
else:
  print("The person is diabetes")
```
